```python
# Importing the necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Reading my supermarket dataset
data=pd.read_csv("/content/supermarket_sales(1).csv")
data.head()
```

```python
# Checking the data type of ech attribute
data.info()
# The info() shows the data type of all the columns that we have in our dataset.
# We don't have any null values present in the data set.
# All the columns that are categorical in nature have object as their data type.
# All the columns that are continuous in nature have int/float as their data type.
```

```python
# Analyzing the statistical fatures using data.describe()
data.describe()
```

```python
# Branch
# Using the value_counts() we get the count of unique values in descending order.
# This shows that we have Branch-A mostly available across different locations followed by B and C.
data['Branch'].value_counts()
```

```python
# Customer Type
# This shows that we have 50% customers have membership and 50% customers don't have a membership
data['Customer type'].value_counts(normalize=True)*100
```

```python
# Gender ratio in our dataset
data['Gender'].value_counts(normalize=True)*100
```

```python
# Product Line
# As per our data supermarket has majority of Fashion accessories 17.8% followed by Food and beverages 17.4%
data['Product line'].value_counts(normalize=True)*100
```

```python
# Payment Method
# Ewallet and Cash are the 1st choice for medium of payment.
data['Payment'].value_counts(normalize=True)*100
```

```python
# Products purchased from a City
b_q=data.groupby(['City','Product line'])['Quantity'].sum().reset_index().sort_values(by='Quantity',ascending=False)
b_q.head()
```

```python
# Bar chart-Which city has the highest products being purchased from a particular product line.
g=px.bar(b_q,x='City',y='Quantity',color='Product line',text='Quantity')
g.show()
```

```python
# Whats the gross income distribution over different branches? 
gross_income=data.groupby(['Branch'])['gross income'].sum().round(2).reset_index().sort_values(by='gross income',ascending=False)
gross_income.head()
```

```python
# No. of products at each line
product_line_report=data.groupby(['Product line'])[['Unit price','Quantity','Tax 5%']].sum().reset_index().round(2)
```

```python
# Creating a new column P_Total that tell us the total price of that category including the tax.
product_line_report['P_Total']=product_line_report['Unit price']*product_line_report['Quantity']+product_line_report['Tax 5%']
```

```python
# Final Report of Product line
product_line_report.head().sort_values(by='P_Total',ascending=False)
```

```python
# Product line
fig=px.line(product_line_report,x='Product line',y='P_Total')
fig.show()
```

```python
# Total=(Unit price*Qunatity)+Tax 5%
# Gross Income=Total-Cost of Goods Sold
# Gross Margin %age=(Total-Cost of Goods Sold)/(Total)
```

```python
# City with highest gross margin percentage
city_report=data.groupby(['City'])[['gross margin percentage']].sum().reset_index().round(2).sort_values(by='gross margin percentage',ascending=False)
city_report.head()
```

```python
# Store that has the highest Rating and the lowest rating
rating=data.groupby(['Branch','Gender','Product line'])[['Rating']].mean().round(2).reset_index().sort_values(by='Rating',ascending=False)
rating.head()
```

```python
# Lowest Rating
rating.tail()
```

```python
# Branch and Tax
payment_tax=data.groupby(['Branch'])[['Tax 5%']].sum().round(2).reset_index().sort_values(by='Tax 5%',ascending=False)
payment_tax.head()
# Branch C has more tax on its products as compared to other A and B.
```

```python
# Tax
fig=px.bar(payment_tax,x='Branch',y='Tax 5%',color='Branch',text_auto=True)
fig.show()
```

```python
# From which gender we have the most amount of profit across each Branch
membership=data.groupby(['Customer type','Gender','Branch'])[['gross margin percentage']].sum().round(2).reset_index().sort_values(by='gross margin percentage',ascending=False)
membership.head(3)
```

```python
# Fuunel 
fig=px.funnel(membership,x='gross margin percentage',y='Customer type',color='Branch')
fig.show()
```

```python
# Gross Income by product line
gross_income=data.groupby('Product line')['gross income'].sum().round(2).reset_index().sort_values(by='gross income',ascending=False)
gross_income.head()
```

```python
# Gross Income-Line Chart
fig=px.line(gross_income,x='Product line',y='gross income',markers=True,text='gross income')
fig.update_traces(textposition='bottom right')
fig.show()
```

```python
# Getting Month from date
import datetime
dt=datetime.datetime.today()
data['Month']=pd.to_datetime(data['Date']).dt.month_name()
```

```python
# Month
month=data.groupby(['Month','Branch'])[['gross income']].sum().reset_index().round(2).sort_values(by=['gross income'],ascending=False)
month.head()
```

```python
# Monthly Gross Income of superstore branches
fig=px.line(month,x='Month',y='gross income',color='Branch',markers=True)
fig.show()
```

```python
# Creating Hour column using Time.
data['Hour']=pd.to_datetime(data['Time']).dt.hour
```

```python
# Hour, Product line and Quantity
time_income=data.groupby(['Hour','Product line'])['Quantity'].sum().reset_index().sort_values(by='Quantity',ascending=False)
time_income.head()
```

```python
# Stacked Bar chart
fig=px.bar(time_income,x='Hour',y='Quantity',color='Product line',text='Quantity')
fig.show()
```

```python
# Gender Based Analysis.
gender=data.groupby(['Gender'])[['Total','gross margin percentage','gross income']].sum().round(2).reset_index().sort_values(by='Total',ascending=False)
gender.head()
# Females seem to contribute more towards the profit of all the supermarkets.
```
